```python
import matplotlib.pyplot as plt
plt.title('plotting') # 그래프의 주목
plt.plot([1,2,3,4],[12,43,25,15])
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title('legend') # 그래프의 주목
plt.plot([10,20,30,40], label='asc')
plt.plot([40,30,20,10], label='desc')
plt.legend() # 레이블 값이 범례로 나타난다.
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title('legend') # 그래프의 주목
plt.plot([10,20,30,40], color='skyblue',label='skyblue')
plt.plot([40,30,20,10], 'pink',label='pink') # color 생략 가능
plt.legend() # 레이블 값이 범례로 나타난다.
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title('linestyle') # 그래프의 주목
plt.plot([10,20,30,40], 'r',linestyle='--',label='dashed')
# linstyle = ls 로 나타낼 수 있습니다.
plt.plot([40,30,20,10], color='g',ls=':', label='dotted')
plt.legend() # 레이블 값이 범례로 나타난다.
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title('linestyle') # 그래프의 주목
plt.plot([10,20,30,40], 'r--', label='dashed')
# linstyle = ls 로 나타낼 수 있습니다.
plt.plot([40,30,20,10], color='g',ls=':', label='dotted')
plt.legend() # 레이블 값이 범례로 나타난다.
plt.show()
```

```python
# 마커 모양 바꾸기
import matplotlib.pyplot as plt
plt.title('marker') # 제목 설정
plt.plot([10,20,30,40], 'r.', label='circle')
plt.plot([40,30,20,10],'g^', label = 'triangle up')
plt.legend() # 범례표시
plt.show()
# plt.plot([10,20,30,40,],'r.--')  위 예시와 같이 로도 가능합니다!
```
