```python
# 히스토그램 예제
import csv
import matplotlib.pyplot as plt
f = open('seoul.csv')
data = csv.reader(f)
next(data)

aug = [] # 8월의 최고 기오 값을 저장
jan = [] # 1월의 최고 기온 값을 저장
for row in data :
    month = row[0].split('-')[1] # 월의 값을 저장  
    if row[-1] !='':                  # 최고 기온 데이터 값이 존재한다면
        if month == '08':
            aug.append(float(row[-1]))
        if month == '01':
            jan.append(float(row[-1]))
# 히스토그램은 color 를 생략이 안된다!
plt.hist(aug, bins= 100,color='r',label ='Aug')
plt.hist(jan, bins= 100,color='b',label ='Jan')        
plt.legend()
plt.show()
```

```python
# 상자 그림 예제
import csv
import matplotlib.pyplot as plt
f = open('seoul.csv')
data = csv.reader(f)
next(data)
day = []
for i in range(31):
    day.append([])

for row in data :
    if row[-1] != '' :
        if row[0].split('-')[1] == '08':
            day[int(row[0].split('-')[2])-1].append(float(row[-1]))

#이 그래프의 스타일로 지정하는 코드            
plt.style.use('ggplot')
# 그래프의 크기를 가로 10 세소 5 로 지정하여 박스의 크키를 수정
plt.figure(figsize=(10,5), dpi = 300)
# 이상치 값이 보이지 않게 설정하는 코드
plt.boxplot(day,showfliers= False)
plt.show()
```
